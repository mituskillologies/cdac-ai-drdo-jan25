```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

##### Simple Histogram

```python
marks = np.array([56,76,89,87,65,61,65,45,69,71,72,42,55,67,81,69,68,61])
```

```python
plt.title('Histogram')
plt.grid()
plt.xlabel('Marks')
plt.xticks(range(0,100,10))
plt.hist(marks, bins=range(0,100,10), color='r');
```

##### Image Histogram

```python
img_gray = cv2.imread('images/sachin.jpg', 0)
```

```python
plt.title('Histogram')
plt.grid()
plt.xlabel('Pixels')
plt.hist(img_gray.flatten(), bins=range(0,256), color='b');
```
