```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/lenna.png', 0)
```

```python
new_img = img > 127
```

```python
new_img
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1);  plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
plt.subplot(1,2,2);  plt.title('Thresholded Image')
plt.imshow(new_img, cmap = 'gray')
```

```python
inv_img = img <= 127
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,3,1);  plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
plt.subplot(1,3,2);  plt.title('Thresholded Image')
plt.imshow(new_img, cmap = 'gray')
plt.subplot(1,3,3);  plt.title('Inverse Thresholded Image')
plt.imshow(inv_img, cmap = 'gray')
```

##### Otsu Thresholding

```python
help(cv2.threshold)
```

```python
otsu_thresh, otsu_img = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
```

```python
otsu_thresh
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1);  plt.title('Original Image')
plt.imshow(img, cmap = 'gray')
plt.subplot(1,4,2);  plt.title('Thresholded Image')
plt.imshow(new_img, cmap = 'gray')
plt.subplot(1,4,3);  plt.title('Inverse Thresholded Image')
plt.imshow(inv_img, cmap = 'gray')
plt.subplot(1,4,4);  plt.title('Otsu Thresholded Image')
plt.imshow(otsu_img, cmap = 'gray')
```
