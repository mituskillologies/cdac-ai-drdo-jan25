##### Resizing using skimage

```python
from skimage.transform import resize
```

```python
img_rgb = plt.imread('images/sachin.jpg')
```

```python
img_new = resize(img_rgb, (450,300))
```

```python
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img_rgb)
plt.subplot(1,2,2)
plt.title('Resized Image')
plt.imshow(img_new)
plt.savefig('Resized.png')
```
