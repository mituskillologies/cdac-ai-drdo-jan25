##### Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

##### Load the data

```python
df = pd.read_csv('datasets/mtcars.csv')
```

```python
df
```

##### Separate the input and output data

```python
# input data
x = df[['disp','hp','wt']]

# output data
y = df['mpg']
```

```python
x
```

```python
import seaborn as sns
```

```python
sns.kdeplot(y);
```

##### Build the model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
x.corrwith(y)
```

```python
disp = 221
hp = 102
wt = 2.91
```

```python
regressor.intercept_ + disp * regressor.coef_[0] + hp * regressor.coef_[1] + wt * regressor.coef_[2]
```

```python
regressor.predict([[disp, hp, wt]])
```
