```python
import pandas as pd
```

```python
dataset = [['Apple','Beer','Rice','Chicken'],
           ['Apple','Beer','Rice'],
           ['Apple','Beer'],
           ['Apple','Pear'],
           ['Milk','Beer','Rice','Chicken'],
           ['Milk','Beer','Rice'],
           ['Milk','Beer'],
           ['Milk','Pear']
          ]
```

```python
dataset
```

##### Transaction Encoding

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
data_new = te.fit_transform(dataset)
```

```python
data_new
```

```python
te.columns_
```

```python
df = pd.DataFrame(data=data_new, columns=te.columns_)
```

```python
df
```

##### 1. Generate the frequent itemsets

```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_itemset = apriori(df, min_support=0.25, use_colnames=True)
```

```python
freq_itemset
```

##### 2. Generate the association rules

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(freq_itemset, min_threshold=0.5, metric='confidence')
```

```python
rules
```

```python
rules = rules.iloc[:,[0,1,4,5]]
```

```python
rules
```

##### Sort Rules

```python
rules.sort_values(by='confidence', ascending=False)
```

```python
rules.sort_values(by='confidence', ascending=False).head(5)
```

```python
rules[rules['confidence'] > 0.6]
```

```python
rules[rules['confidence'] > 0.6].sort_values(by='confidence', ascending=False)
```

##### Recomendations

```python
rules[rules['antecedents'] == {'Apple'}]
```

```python
rules[rules['antecedents'] == {'Apple','Beer'}]['consequents']
```

```python
rules[rules['antecedents'] == {'Beer','Apple'}]['consequents']
```

##### Export the rules

```python
rules.to_csv('rules.csv', index=False)
```
