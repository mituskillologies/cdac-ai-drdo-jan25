##### Bar plot

```python
studs = [67,79,156,189,104,111,58]
langs = ['C','C++','Java','Python','PHP','JavaScript','R']
```

```python
plt.title('Student Enrollment')
plt.xlabel('Course')
plt.ylabel('Total Students')
plt.bar(langs, studs, color = 'm')
```

##### Horizontal Bar Chart

```python
plt.title('Student Enrollment')
plt.ylabel('Course')
plt.xlabel('Total Students')
plt.barh(langs, studs, color = 'pink')
```

```python
plt.title('Student Enrollment')
plt.ylabel('Student Name')
plt.xlabel('Marks')
plt.barh(student['name'], student['marks'], color = 'navy')
```

##### Stacked Bar Chart

```python
plt.title('Student Enrollment')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(student['name'], student['marks']+student['age'], color = 'yellow',
       label = 'age')
plt.bar(student['name'], student['marks'], color = 'blue', label = 'marks')
plt.legend()
```

```python
plt.title('Student Enrollment')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(student['name'], student['marks'], color = 'orange', label = 'marks',
       align='center')
plt.bar(student['name'], student['age'], color = 'cyan', label = 'age', 
        align='edge', width=0.5)
plt.legend()
```

##### Pie Chart

```python
plt.figure(figsize=(10,10))
plt.pie(studs, labels=langs, autopct='%2.2f%%');
```
