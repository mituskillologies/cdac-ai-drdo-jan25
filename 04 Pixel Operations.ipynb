```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros([200,200], dtype=np.uint8)
```

```python
img
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
img[100,100] = 255
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
img[100,:] = 255
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
for i in range(200):
    img[i,i] = 255
```

```python
plt.imshow(img, cmap = 'gray')
```

##### Saving the images

```python
cv2.imwrite('myimage.jpg', img)
```

```python
plt.imsave('myimage1.jpg', img)
```

```python
from skimage import io
```

```python
io.imsave('myimage2.jpg', img)
```

```python
img_cv = cv2.imread('images/sachin.jpg')
img_plt = plt.imread('images/sachin.jpg')
```
