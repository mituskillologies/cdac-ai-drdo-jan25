```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array([5,9,12,11,3,6,4])
```

```python
def scaler(n):
    x_new = (n - n.min()) / (n.max() - n.min())
    return x_new
```

```python
scaler(x)
```

```python
y = np.array([506, 788, 934, 476, 178, 345, 643, 457])
```

```python
scaler(y)
```

```python
# Dataset: Social_Network_Ads.csv
# Location: https://mitu.co.in/dataset
```

```python
df = pd.read_csv('datasets/Social_Network_Ads.csv')
```

```python
df
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(df['Age'], df['EstimatedSalary'], c = df['Purchased'])
```

```python
df.describe()
```

```python
x = df[['Age','EstimatedSalary']]
```

```python
# import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# create the object
scaler = MinMaxScaler()
```

```python
# apply the method
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled = pd.DataFrame(data=x_scaled, columns=x.columns)
```

```python
x_scaled
```

```python
x_scaled.describe()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Original Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(df['Age'], df['EstimatedSalary'], c = df['Purchased'])

plt.subplot(1,2,2)
plt.title('Scaled Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(x_scaled['Age'], x_scaled['EstimatedSalary'], c = df['Purchased'])
```
