```python
df = pd.read_csv('datasets/ingredients.csv')
```

```python
sns.jointplot(x='sweetness', y='crunchiness', hue='class', data=df)
```

```python
plt.xlim(0,12); plt.ylim(0,12)
plt.scatter(df['sweetness'], df['crunchiness'])
for i in range(15):
    plt.annotate(text=df['ingredients'][i], xy=(df['sweetness'][i], df['crunchiness'][i]))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
x = df[['sweetness','crunchiness']]
y = df['class']
```

```python
knn.fit(x, y)
```

```python
tomato = pd.DataFrame(data=[[6,4]], columns=x.columns)
tomato
```
