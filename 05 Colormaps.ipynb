```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sachin.jpg', 0)
```

##### Sequential

```python
plt.imshow(img, plt.colormaps['Greys'].reversed())
```

```python
plt.imshow(img, plt.colormaps['Blues'].reversed())
```
