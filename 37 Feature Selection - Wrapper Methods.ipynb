```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('datasets/mtcars.csv')
```

```python
x = df[['disp','hp','wt']]
y = df['mpg']
```

```python
reg = LinearRegression()
reg.fit(x, y)
print('Accuracy:', reg.score(x, y))
```

```python
x = df.drop('mpg', axis = 1)
```

##### Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(n_features_to_select=3, estimator=LinearRegression() )
```

```python
rfe.fit(x, y)
```

```python
rfe.get_support()
```

```python
x_new = x.iloc[:, rfe.get_support()]
x_new
```

```python
reg = LinearRegression()
reg.fit(x_new, y)
print('Accuracy:', reg.score(x_new, y))
```
