```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread('images/sachin.jpg')
```

```python
pyramids = [img]
```

```python
pyramids
```

```python
len(pyramids)
```

```python
for i in range(5):
    img = cv2.pyrDown(img)
    pyramids.append(img)
```

```python
len(pyramids)
```

```python
plt.figure(figsize=(20,5))
for i in range(6):
    plt.subplot(1,6,i+1)
    plt.imshow(pyramids[i])
```

```python
for i in range(6):
    plt.imsave('sachin%d.jpg' %i, pyramids[i])
```
