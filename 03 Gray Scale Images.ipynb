```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sachin.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('Gray Image', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
